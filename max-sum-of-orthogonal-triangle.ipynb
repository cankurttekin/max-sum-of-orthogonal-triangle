```python
def fileToArr(path):
  rows = []
  file = open(path)
  for line in file:
    rows.append(line.split())
    
  # str arr elements to int
  for i in range(len(rows)):
    for j in range(len(rows[i])):
      rows[i][j] = int(rows[i][j])
  return rows
```

```python
def isPrime(x):

  # if the number is 2 its prime
  if x == 2:
    return 1

  # if the number is even or less than 2, not prime
  if x % 2 == 0 or x < 2 :
    return 0
    
  # division check up to the square root of the number
  i=3
  while i <= x ** (1/2):
    if x % i == 0:
      return 0
    i += 2 # if the number is even or not checked in the beginning so no need to try to divide to even
 
  return 1
```

```python
def primesAndPathBlock(arr):
  for i in range(len(arr) - 2, -1, -1):
      for j in range(len(arr[i])):
        if isPrime(arr[i][j]) :
          arr[i][j] = -999
        if isPrime(arr[i+1][j]) and isPrime(arr[i+1][j+1]):
          arr[i+1][j] = -888
          arr[i+1][j+1] = -888
          arr[i][j] = -777
  return arr
```

```python
def maxSumOfOrthogonalTriangle(arr):
    for i in range(len(arr) - 2, -1, -1):
      for j in range(len(arr[i])):
        if arr[i][j]>0 and (arr[i+1][j] > 0 or arr[i+1][j+1] > 0):
          arr[i][j] += max(arr[i+1][j],arr[i+1][j+1])
    return arr[0][0]
```
